```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Extract

```python
df = pd.read_csv('flavors_of_cacao.csv', sep=',')
df
```

Transform

```python
df.columns = ['Company', 'Bean_Specific_Region', 'Insertion_Database', 'Review_Year', 'Cocoa_Percentage', 'Company_Location','Rating','Bean_Type','Bean_Broad_Region']
df.head()
```

```python
df.dtypes
```

```python
df['Cocoa_Percentage']= df['Cocoa_Percentage'].str.replace(r'\D','')
df
```

```python
df['Cocoa_Percentage'] = df['Cocoa_Percentage'].astype(float)
df.dtypes
```

```python
df.loc[df['Cocoa_Percentage'] > 100]
```

```python
df = df.drop(df[df['Cocoa_Percentage'] > 100].index)
df.loc[df['Cocoa_Percentage'] > 100]
```

```python
df.isnull().sum()
```

```python
df['Bean_Broad_Region']= df['Bean_Broad_Region'].replace(r'^\s*$', np.nan, regex=True)
df.isna().sum()
```

```python
df['Bean_Type']= df['Bean_Type'].replace(r'^\s*$', np.nan, regex=True)
df.isna().sum()
```

```python
df = df.drop('Bean_Type', axis='columns')
df
```

```python
df = df.dropna()
df.isna().sum()
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.describe
```

```python
companies_ratings = df.groupby('Company',as_index=False).agg(
    average_rating = ('Rating', 'mean'),
    max_rating = ('Rating', 'max'),
    min_rating = ('Rating','min')
).round(2).sort_values(by='average_rating', ascending=False)

companies_ratings
```

```python
best_cocoa_ratings = df.groupby('Bean_Specific_Region',as_index=False).agg(
    number_entries = ('Bean_Specific_Region', 'count'),
    max_rating = ('Rating', 'max'),
    average_rating =  ('Rating', 'mean')
    ).round(2).sort_values(by='average_rating', ascending=False)

best_cocoa_ratings
```

```python
countries_best_ratings = df.groupby('Bean_Broad_Region',as_index=False).agg(
    number_entries = ('Bean_Broad_Region', 'count'),
    max_rating = ('Rating', 'max'),
    average_rating =  ('Rating', 'mean')
    ).round(2).sort_values(by='average_rating', ascending=False)

countries_best_ratings
```

```python
cocoa_percentage_ratings = df.groupby('Cocoa_Percentage',as_index=False).agg(
    number_entries = ('Cocoa_Percentage', 'count'),
    max_rating = ('Rating', 'max'),
    min_rating = ('Rating', 'min'),
    average_rating =  ('Rating', 'mean')
    ).round(2).sort_values(by='average_rating', ascending=False)

cocoa_percentage_ratings
```

Load

```python
cocoa_percentage_ratings.to_csv('/cocoa_percentage_ratings.csv',index=False)
```

```python
best_cocoa_ratings.to_csv('/best_cocoa_ratings.csv',index=False)
```

```python
companies_ratings.to_csv('/companies_ratings.csv',index=False)
```

```python
countries_best_ratings.to_csv('/countries_best_ratings.csv',index=False)
```
